# Suppose that a hospital tested the age and body fat data for 18 randomly selected adults with the following results. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

dataset = pd.read_csv(r"C:\Users\saket\OneDrive\Desktop\ML\Datasets\hospital_age_body_fat_data.csv")

# Display the first few rows
print("Dataset:")
print(dataset.head(), "\n")
```

# a) Find the maximum and minimum percentage of fat and age

```python
max_fat = dataset['Body Fat (%)'].max()
min_fat = dataset['Body Fat (%)'].min()
max_age = dataset['Age'].max()
min_age = dataset['Age'].min()

print(f"Maximum Body Fat (%): {max_fat}")
print(f"Minimum Body Fat (%): {min_fat}")
print(f"Maximum Age: {max_age}")
print(f"Minimum Age: {min_age}\n")
```

# b) Calculate mean, median, and midrange of the age

```python
mean_age = dataset['Age'].mean()
median_age = dataset['Age'].median()
midrange_age = (max_age + min_age) / 2

print(f"Mean Age: {mean_age}")
print(f"Median Age: {median_age}")
print(f"Midrange Age: {midrange_age}\n")
```

# c) Find the first quartile and third quartile of the data

```python
first_quartile_age = dataset['Age'].quantile(0.25)
third_quartile_age = dataset['Age'].quantile(0.75)

print(f"First Quartile (Q1) Age: {first_quartile_age}")
print(f"Third Quartile (Q3) Age: {third_quartile_age}\n")
```

# d) Draw a scatter plot and Q-Q plot based on these two variables

```python
plt.figure(figsize=(14, 6))

# Scatter plot
plt.subplot(1, 2, 1)
sns.scatterplot(x='Age', y='Body Fat (%)', data=dataset)
plt.title('Scatter Plot of Age vs. Body Fat (%)')
plt.xlabel('Age')
plt.ylabel('Body Fat (%)')

# Q-Q plot
plt.subplot(1, 2, 2)
stats.probplot(dataset['Body Fat (%)'], dist="norm", plot=plt)
plt.title('Q-Q Plot of Body Fat (%)')

plt.tight_layout()
plt.show()
```
