# 1. Given a data set “data” for analysis it includes the attribute Country, purchased item, age, Salary. 

```python
import numpy as np
import pandas as pd

dataset = pd.read_csv(r"C:\Users\saket\OneDrive\Desktop\ML\Datasets\converted_data.csv")
dataset
```

# a. Identify number of missing values in a given dataset 

```python
dataset.isnull().any()
```

# b. Drop the tuples that have missing values in the attributes. 

```python
df = dataset.dropna(axis=0)
```

```python
df
```

# c. Check the data type of age , ifit is not an integer then convert into integer. 

```python
df['Age'] = df['Age'].astype(np.int32())
```

```python
df
```

# d. Normalize the salary using simple feature scaling. 

```python
df['salary_normalized'] = (df['Salary'] - df['Salary'].min()) / (df['Salary'].max() - df['Salary'].min())
```

```python
df
```

# e. Categorize the salary into low, high, medium bins. 

```python
bins = [df['Salary'].min(), df['Salary'].quantile(0.33), df['Salary'].quantile(0.66), df['Salary'].max()]
labels = ['Low', 'Medium', 'High']

df['salary_category'] = pd.cut(df['Salary'], bins=bins, labels=labels, include_lowest=True)
```

```python
df
```

# f. Turn the categorical values into numerical.

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

df['salary_category_numeric'] = label_encoder.fit_transform(df['salary_category'])
```

```python
df = pd.get_dummies(df, columns=['salary_category'])
```

```python
df
```
